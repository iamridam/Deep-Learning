```python
import numpy as np
import pandas as pd
```

```python
# %env CUDA_DEVICE_ORDER = PCI_BUS_ID
# %env CUDA_VISIBLE_DEVICES = 0,1
```

```python
alphabet = list("ACDEFGHIKLMNPQRSTVWY")
```

```python
len(alphabet)
```

```python
train_data = pd.read_csv("trainset.data")
```

```python
train_data.info()
```

```python
train_data.head()
```

```python
X_train = []

for sequence, label in train_data.values:
    frequency = {}
    for letter in alphabet:
        frequency[letter] = 0
    for letter in sequence:
        frequency[letter] = (frequency[letter] * len(sequence) + 1) / len(sequence)
    X_train.append(list(frequency.values()))
```

```python
X_train_df_1 = pd.DataFrame(X_train, columns=alphabet)
```

```python
X_train_df_1.head()
```

```python
X_train_df = pd.concat([X_train_df_1, train_data[['Label']]], axis=1)
```

```python
bipeptide = pd.read_csv('train_seq_dpv_1_online.csv')
```

```python
bipeptide.head()
```

```python
bipeptide.drop(['ID'], inplace = True, axis = 1)
```

```python
bipeptide.shape
```

```python
tripeptide = pd.read_csv('tripeptide.csv')
```

```python
tripeptide.shape
```

```python
X_train_df.shape
```

```python
X_train_df = pd.concat([tripeptide, bipeptide, X_train_df,], axis=1)
```

```python
X_train_df.head()
```

```python
#Dependencies
import keras
from keras.models import Sequential
from keras.layers import Dense

# Neural network
model = Sequential()
model.add(Dense(1024, input_dim=8420, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
X_train = X_train_df.iloc[:, :-1].values
y_train = X_train_df.iloc[:, -1].values
```

```python
for index, label in enumerate(y_train):
    if label == -1:
        y_train[index] = 0
```

```python
y_train
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
history = model.fit(X_train, y_train, epochs=25, batch_size=64)
```

```python
X_test = []

for _id, sequence in test_data.values:
    frequency = {}
    for letter in alphabet:
        frequency[letter] = 0
    for letter in sequence:
        frequency[letter] = (frequency[letter] * len(sequence) + 1) / len(sequence)
    X_test.append(list(frequency.values()))
```
